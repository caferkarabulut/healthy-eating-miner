```python
numeric_cols = [
    "calories",
    "protein_g",
    "carbs_g",
    "fat_g",
    "fiber_g",
    "sugar_g",
    "sodium_mg",
    "cholesterol_mg"
]
```

```python
import pandas as pd

# Yeni kategorik sütunları oluştur
for col in numeric_cols:
    df[col + "_cat"] = pd.qcut(df[col], q=3, labels=["low", "medium", "high"])
```

```python
df[[col + "_cat" for col in numeric_cols]].head()
```

```python
from mlxtend.preprocessing import TransactionEncoder

# Apriori için kategorik etiketleri birleştirelim
basket = []

for idx, row in df.iterrows():
    items = []
    for col in numeric_cols:
        items.append(f"{col}_cat=" + str(row[col + "_cat"]))
    items.append("cuisine=" + row["cuisine"])
    items.append("meal_type=" + row["meal_type"])
    items.append("diet_type=" + row["diet_type"])
    items.append("cooking_method=" + row["cooking_method"])
    basket.append(items)

# One-hot encoding
te = TransactionEncoder()
te_ary = te.fit(basket).transform(basket)
df_apriori = pd.DataFrame(te_ary, columns=te.columns_)
df_apriori.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemsets = apriori(df_apriori, 
                            min_support=0.05, 
                            use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,
                          metric="lift",
                          min_threshold=1.0)

rules = rules.sort_values("lift", ascending=False)
rules.head(20)
```

```python
# En güçlü 10 kural
top_rules = rules.head(10).copy()
top_rules
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Kural metnini string'e çeviriyoruz (antecedent -> consequent)
rule_labels = [
    f"{list(row['antecedents'])[0]} → {list(row['consequents'])[0]}"
    for _, row in top_rules.iterrows()
]

plt.barh(range(len(top_rules)), top_rules["lift"])
plt.yticks(range(len(top_rules)), rule_labels)
plt.xlabel("Lift")
plt.title("En Güçlü 10 Birliktelik Kuralı (Lift Değerine Göre)")
plt.gca().invert_yaxis()  # En yüksek lift en üstte olsun
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(rules["support"], rules["confidence"])
plt.xlabel("Support (Destek)")
plt.ylabel("Confidence (Güven)")
plt.title("Tüm Kuralların Destek–Güven Dağılımı")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top20 = rules.head(20)

plt.figure(figsize=(8, 6))
plt.scatter(top20["support"], top20["confidence"])
for i, row in top20.iterrows():
    label = f"{list(row['antecedents'])[0]} → {list(row['consequents'])[0]}"
    plt.annotate(label, (row["support"], row["confidence"]), fontsize=7, xytext=(2,2), textcoords="offset points")

plt.xlabel("Support (Destek)")
plt.ylabel("Confidence (Güven)")
plt.title("En Güçlü 20 Kuralın Destek–Güven Dağılımı")
plt.grid(True)
plt.tight_layout()
plt.show()
```
